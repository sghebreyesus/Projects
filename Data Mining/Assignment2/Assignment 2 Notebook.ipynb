```python
pd.set_option("display.max_columns", 100)
url = "https://raw.githubusercontent.com/cs6220/cs6220.spring2019/master/data/AmesHousing.txt"
df_ames = pd.read_csv(url, sep="\t")
print(type(df_ames))
```

```python
#Question 2.1.1 - This method tells you the number of columns that contain a NaN value. There are 27 features that have
#missing values.
df_ames.isnull().any(axis=0).sum()
```

```python
#Question 2.1.2 - Created a list of columns that are nominal features and contain missing values. I included ordinal
#features in the list of nominal features as well (it was easier to take them as nominal features rather than numeric
#features). Then, I used .fillna to fill those nominal columns with the string 'missing'.  
nominal_cols = ['Alley', 'Mas Vnr Type', 'Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2', 
        'Electrical', 'Fireplace Qu', 'Garage Type', 'Garage Finish', 'Garage Qual', 'Garage Cond', 'Pool QC', 'Fence', 
        'Misc Feature']
df_ames[nominal_cols] = df_ames[nominal_cols].fillna('missing')
df_ames.head()
```

```python
#Question 2.1.3 - Created a list of columns that are numeric features and contain missing values. Then, I interpolated
#those missing numeric features using linear interpolation. 
numeric_cols = ['Lot Frontage', 'Mas Vnr Area', 'BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF',
               'Bsmt Full Bath', 'Bsmt Half Bath', 'Garage Yr Blt', 'Garage Cars', 'Garage Area']
df_ames[numeric_cols] = df_ames[numeric_cols].interpolate()
df_ames.head()
```

```python
#Question 2.2.1 - Standardization. I only standardized numeric features, except Order Number, PID and SalePrice. I created
#a list of columns that were considered numeric features and performed z-score standardization. 
from sklearn.preprocessing import StandardScaler
numer_cols = ['Lot Frontage', 'Lot Area', 'Overall Qual', 'Overall Cond', 'Year Built', 'Year Remod/Add', 'Mas Vnr Area', 
             'BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF', '1st Flr SF', '2nd Flr SF', 
             'Low Qual Fin SF', 'Gr Liv Area', 'Bsmt Full Bath', 'Bsmt Half Bath', 'Full Bath', 'Half Bath', 
             'Bedroom AbvGr', 'Kitchen AbvGr', 'TotRms AbvGrd', 'Fireplaces', 'Garage Yr Blt', 'Garage Cars', 'Garage Area', 
             'Wood Deck SF', 'Open Porch SF', 'Enclosed Porch', '3Ssn Porch', 'Screen Porch', 'Pool Area', 'Misc Val', 
             'Mo Sold', 'Yr Sold']
scaler = StandardScaler()
scaler.fit(df_ames[numer_cols])
scaled = scaler.fit_transform(df_ames[numer_cols])
scaled_df = pd.DataFrame(scaled, columns=numer_cols)
scaled_df.head()
```

```python
#Question 2.2.2 - Boxplot before standardization. This boxplot has a much wider range of possible values than those made
#after standardization, as we see that the y-axis scale goes all the way up to 200,000. The feature with the furthest outlier
#seems to be Lot Area. 
df_ames[numer_cols].boxplot(grid=False, figsize = (100, 100), return_type='axes')
```

```python
#Question 2.2.2 - Boxplot after standardization. This boxplot has a much narrower range of values compared to the ones
#made before feature standardization, as we can see the y-axis scale only goes up to 30. The feature with the furthest
#outlier seems to be misc val. 
scaled_df.boxplot(grid=False, figsize = (100, 100), return_type='axes')
```

```python
#Added SalePrice to scaled_df dataframe. SalePrice is not standardized, as it is the target variable
scaled_df['SalePrice'] = df_ames['SalePrice']
scaled_df.head()
```

```python
#Question 2.3.1 - Mutual Information 
from sklearn.feature_selection import mutual_info_regression
from sklearn.feature_selection import SelectKBest
from sklearn.datasets import make_swiss_roll
X = scaled_df.iloc[:, :-1]
y = scaled_df.iloc[:, -1]
mi = mutual_info_regression(X, y)
print(mi)
```

```python
#Question 2.3.2 - Top 5 numeric features 
selector = SelectKBest(mutual_info_regression, k=5)
selector.fit(X, y)
idxs_selected = selector.get_support(indices=True)
X_new = X.iloc[:, idxs_selected]
print(X_new.head())
```

```python
#Question 2.3.3 - First found the top 1 feature, which I found to be Overall Qual (Rates the overall material and finish of 
#the house between 1-10). I think the higher this value is, the higher the sale price will be because a higher rating means
#the house is of higher quality (may have more amenities, more space, better quality materials, less renovations needed, etc.) 
#and a higher quality house will fetch higher prices on the market as there is more demand for them and limited supply. 
selector = SelectKBest(mutual_info_regression, k=1)
selector.fit(X, y)
idxs_selected = selector.get_support(indices=True)
X_new = X.iloc[:, idxs_selected]
print(X_new.head())
```
